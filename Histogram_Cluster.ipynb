```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

#### the column 'visN' contains measurement numbers for CHLORIDE LEVEL

```python
data=pd.read_csv('histogram_cluster.csv')
data
```

# Histogram

```python
import matplotlib.pyplot as plt
```

```python
bin_range=32
bins=np.arange(bin_range) - 0.5
plt.hist(data['visN'],bins=bins, alpha=0.3, color='r')
```

```python
A_bool=np.logical_or(data['visN']==1, data['visN']==2)
B_bool=np.logical_or(data['visN']==6, data['visN']==7)
B_bool=np.logical_or(B_bool, data['visN']==8)
```

```python
AB_bool=np.logical_or(A_bool, B_bool)
```

```python
data['subgroup']='else'
data['subgroup'][A_bool]='A'
data['subgroup'][B_bool]='B'
```

```python
ABgroup=data[AB_bool]
ABgroup
```

# TSNE clustering - HypothesisII with top 15 variables

```python
from sklearn.manifold import TSNE
import seaborn as sns
palette=sns.color_palette('bright',2)

redux_ver1=TSNE(n_components=2, perplexity=50).fit_transform(ABgroup.iloc[:,0:15])
```

```python
sns.scatterplot(redux_ver1[:,0],redux_ver1[:,1],hue=ABgroup['subgroup'], palette=['blue','green'])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

# Assign Group1 and Group 2 by GMM

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
cn=2
gm_kmeans = GaussianMixture(n_components=cn, 
                    max_iter=1000, 
                    tol=1e-4,
                    init_params='kmeans').fit(data.iloc[:,:-2])
labels=gm_kmeans.predict(data.iloc[:,:-2])
```

```python
from sklearn.manifold import TSNE
import seaborn as sns
color=['blue','green']

redux_ver1=TSNE(n_components=2, perplexity=50).fit_transform(data.iloc[:,:-2])

sns.scatterplot(redux_ver1[:,0],redux_ver1[:,1],hue=labels, palette=color[:cn])

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
